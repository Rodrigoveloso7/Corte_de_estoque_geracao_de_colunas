# PROBLEMA PRINCIPAL

```python
def problema_principal(padroes,demanda):
    #Declaração do modelo
    modelo_principal = ConcreteModel()
    #Definição de variáveis de interesse
    modelo_principal.indices = range(len(padroes))
    modelo_principal.indices_demanda = range(len(demanda))
    modelo_principal.x = Var(modelo_principal.indices,domain=NonNegativeReals)

    #Definição da função objetivo
    modelo_principal.f_objetivo = Objective(expr=sum(modelo_principal.x[i] for i in modelo_principal.indices), sense=minimize)
    print(modelo_principal.f_objetivo.expr)

    #Restrições
    modelo_principal.restr = ConstraintList()
    for j in modelo_principal.indices_demanda:
        expr = sum(modelo_principal.x[i]*padroes[i][j] for i in modelo_principal.indices)
        modelo_principal.restr.add(expr >= demanda[j])
        print(f"Restrição {j}: {expr} >= {demanda[j]}")
    modelo_principal.dual = Suffix(direction=Suffix.IMPORT)
    resultado_principal = SolverFactory('cbc', executable='/usr/bin/cbc').solve(modelo_principal,tee=True)
    variaveis_duais = []
    for i in range(1,len(demanda)+1):
        variaveis_duais.append(modelo_principal.dual[modelo_principal.restr[i]])
    variaveis_de_interesse_principal = []
    for i in modelo_principal.indices:
        print(value(modelo_principal.x[i]))
        variaveis_de_interesse_principal.append(value(modelo_principal.x[i]))
    print("Variáveis duais",variaveis_duais)
    return variaveis_de_interesse_principal,variaveis_duais
```

# PROBLEMA DA MOCHILA

```python
def problema_mochila(padroes,capacidade,variaveis_duais):
    # Declaração do modelo
    modelo_mochila = ConcreteModel()
    # Definição de variáveis de interesse
    modelo_mochila.indices = range(len(variaveis_duais))
    print("padrões",padroes)
    modelo_mochila.x = Var(modelo_mochila.indices, within=NonNegativeIntegers)
    # Definição da função objetivo
    modelo_mochila.f_objetivo = Objective(expr=sum(modelo_mochila.x[i]*variaveis_duais[i] for i in modelo_mochila.indices),
                                            sense=maximize)
    modelo_mochila.rest = Constraint(expr=sum(modelo_mochila.x[i]*comp_itens[i] for i in modelo_mochila.indices )<= capacidade)
    resultado_mochila = SolverFactory('cbc', executable='/usr/bin/cbc').solve(
        modelo_mochila)
    novo_padrao = []
    for i in modelo_mochila.indices:
        novo_padrao.append(value(modelo_mochila.x[i]))
    return novo_padrao,variaveis_duais,modelo_mochila.indices
```

# SOLUCIONADOR

```python
def solucionador(comp_itens,demanda,capacidade,max_iter,padroes):
    padroes = np.eye(len(comp_itens)).tolist()  # verificar depois se inicializar
    # com matrizes menor é melhor
    print(padroes)
    for z in range(max_iter):
        solucao_principal = problema_principal(padroes,demanda)
        solucao_mochila = problema_mochila(padroes,capacidade,solucao_principal[1])
        if sum(solucao_mochila[1][i] * solucao_mochila[0][i] for i in solucao_mochila[2]) > 1:
            print(solucao_mochila[0])
            padroes.append(solucao_mochila[0])
        else:
            print('Solução ótima atingida')
            return padroes, solucao_principal[0]
    # If max_iter is reached, return the current patterns and solution
    return padroes, solucao_principal[0]
```

# TRATAMENTO DE RESÍDUOS

```python
def tratamento_residuos(padroes_corte,quantidade_cortada_relaxada,demanda):
  lista_padroes_corte = np.array(padroes_corte)
  lista_quantidade_cortada_relaxada = np.array(quantidade_cortada_relaxada)
  lista_padroes_corte_att = lista_padroes_corte.copy()
  lista_quantidade_cortada_att = lista_quantidade_cortada_relaxada.copy()
  demanda_att = demanda.copy()
  contador_trat = 1
  # Obtendo indíces para ordenar as listas de padroes e quantidades em ordem decrescente
  indices_decrescentes = np.argsort(lista_quantidade_cortada_relaxada)[::-1]
  def qte_inteira(ind):
      qte_arr = (np.ceil(lista_quantidade_cortada_relaxada[ind]))
      qte_inteira = lista_padroes_corte[ind]* qte_arr
      return qte_inteira,qte_arr
  def qte_inteira_baixo(ind):
      if lista_quantidade_cortada_att[ind] != int(lista_quantidade_cortada_att[ind]):
        qte_arr_baixo = np.floor(lista_quantidade_cortada_att[ind])
      else:
        qte_arr_baixo = lista_quantidade_cortada_att[ind]-1
      qte_inteira = lista_padroes_corte_att[ind]*qte_arr_baixo
      return qte_inteira,qte_arr_baixo
  for ind in indices_decrescentes:
    qte_inteira_res,qte_att = qte_inteira(ind)
    teste_demanda = qte_inteira_res > demanda_att
    if teste_demanda.any():
      while (teste_demanda.any()) and (contador_trat<100):
        qte_inteira_res,qte_att = qte_inteira_baixo(ind)
        contador_trat +=1
        lista_quantidade_cortada_att[ind] = qte_inteira_baixo(ind)[1]
        # print('x',qte_inteira_res)

        teste_demanda = qte_inteira_res > demanda_att
        print('qte inteira res',qte_inteira_res)
        print('demanda att',demanda_att)
        print('teste demandaaa',teste_demanda)
    demanda_att = demanda_att - qte_inteira_res
    print('Demandaaaaaaaaaaaa',demanda_att)
    lista_quantidade_cortada_att[ind] = qte_att
    print('Qteeeeeeeeeee',lista_quantidade_cortada_att)
  ('saiuuuuuuuuuuuuuuuuuuuuuuuuuuu')
  return lista_padroes_corte_att,lista_quantidade_cortada_att,demanda_att
```

# INICIALIZADOR

```python
def inicializador(comp_itens, demanda, capacidade, max_iter, padroes_entrada=None, limite_residual=23):
    if padroes_entrada is None:
        padroes_entrada = np.eye(len(comp_itens)).tolist()

    # Primeira iteração normal
    padroes_corte, quantidade_cortada_relaxada = calculador(
        comp_itens, demanda, capacidade, max_iter, padroes_entrada
)
    lista_padroes_corte_att, lista_quantidade_cortada_att, demanda_att = tratamento_residuos(
        padroes_corte, quantidade_cortada_relaxada, demanda
    )

    # histórico
    lista_padroes_corte_final = np.array(lista_padroes_corte_att, dtype=int)
    lista_quantidade_cortada_final = np.array(lista_quantidade_cortada_att, dtype=int)
    demanda_final = demanda_att.copy()

    contador = 1
    maximo = 5
    print('ITERAÇÃO vNORMALLLLLLLLLLLLLLLLLLLLL',demanda_att)
    while (demanda_att > 0).any() and (contador < maximo):
        print('ITERAÇÃO scundariaaaaaaaaaaaaaaaaaaaaaaaaa',demanda_att)
        print('DEMANDAAAAAAAAAAAAA',demanda_att.sum())
        if demanda_att.sum() <= limite_residual:
            # =====================================================
            # Heurística gulosa da demanda residual
            # =====================================================
            padroes_corte = []
            quantidade_cortada_relaxada = []

            demanda_temp = demanda_att.copy()

            while demanda_temp.sum() > 0:
                padrao = [0] * len(comp_itens)
                espaco = capacidade

                # escolhe itens por ordem da razão demanda[i]/comp_itens[i]
                indices_ord = sorted(
                    range(len(comp_itens)),
                    key=lambda i: (demanda_temp[i] / comp_itens[i]) if comp_itens[i] > 0 else 0,
                    reverse=True,
                )

                for i in indices_ord:
                    if demanda_temp[i] <= 0 or comp_itens[i] > espaco:
                        continue
                    # máximo que cabe da peça i
                    qtd = min(demanda_temp[i], espaco // comp_itens[i])
                    if qtd > 0:
                        padrao[i] = qtd
                        demanda_temp[i] -= qtd
                        espaco -= qtd * comp_itens[i]

                if sum(padrao) > 0:
                    padroes_corte.append(padrao)
                    quantidade_cortada_relaxada.append(1)  # cada padrão uma vez
                else:
                    break  # não cabe mais nada

        else:
            # =====================================================
            # Solver normal (subproblema mochila)
            # =====================================================
            print('Iteraçao padraooooooo')
            print('Demandaaaaaaaaaaaaaaaaaaaaa_ATTTTTTTTTTTTTT',demanda_att.sum())
            padroes_corte, quantidade_cortada_relaxada = calculador(
                comp_itens, demanda_att, capacidade, max_iter, padroes_corte
            )

        # aplica tratamento dos resíduos
        lista_padroes_corte_att, lista_quantidade_cortada_att, demanda_att = tratamento_residuos(
            padroes_corte, quantidade_cortada_relaxada, demanda_att
        )

        # acumula
        lista_padroes_corte_final = np.append(
            lista_padroes_corte_final, lista_padroes_corte_att, axis=0
        )
        lista_quantidade_cortada_final = np.append(
            lista_quantidade_cortada_final, lista_quantidade_cortada_att, axis=0
        )

        contador += 1

    return lista_padroes_corte_final, lista_quantidade_cortada_final, demanda_final
```

# RESOLUÇÃO

```python
lista_padroes_corte_final, lista_quantidade_cortada_final, demanda_final = inicializador(comp_itens,demanda,capacidade,max_iter)
```
